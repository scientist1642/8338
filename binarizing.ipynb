```python
import numpy as np
from PIL import Image
import pandas as pd

import matplotlib.pyplot as plt
 
# Setting the points for cropped image

im = Image.open('images/hand.png')
width, height = im.size
width, height = int(width/2.26), int(height/2.26)
k = 1.628
width, height = int(width/k), int(height/k)
im = im.resize((width, height))
im = np.array(im.convert('L'))
#print(type(im_gray))

xx = im < 128
plt.imshow(xx)
print (np.sum(xx))

ap = np.array(np.where(xx==True)).T
ap[:,0] -= min(ap[:,0])
ap[:,1] -= min(ap[:,1])
ap = ap[:8338]
print (np.shape(ap))
```

```python
# Save it to data
np.savetxt('data/data.csv', ap.astype(int), delimiter=',', header='y,x', comments='', fmt='%i') 
```
